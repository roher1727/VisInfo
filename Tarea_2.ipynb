```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
rendimiento = pd.read_csv("datos.csv")
```

```python
rendimiento
```

```python
rendimiento["Speedup"] = [rendimiento["Time [s]"][0] / t for t in rendimiento["Time [s]"].values]
rendimiento["Efficiency"] = [s / p for p, s in zip(rendimiento["Proc."].values, rendimiento["Speedup"].values)]
rendimiento["Cost"] = [p * t for t, p in zip(rendimiento["Time [s]"].values, rendimiento["Proc."].values)]
```

```python
rendimiento
```

## Graficaci√≥n de valores obtenidos

#### Speedup, Efficiency, Cost

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
fig.set_size_inches(19, 7)
                    
ax1.set_title("Speedup")
ax1.plot(rendimiento["Proc."], rendimiento["Speedup"], marker='o', mec = 'r',linestyle='-')
ax1.set_xlabel("Processors")
ax1.set_ylabel("Speedup")
ax1.grid()

ax2.set_title("Efficiency")
ax2.plot(rendimiento["Proc."], rendimiento["Efficiency"], marker='o', mec = 'r',linestyle='-')
ax2.set_xlabel("Processors")
ax2.set_ylabel("Efficiency")
ax2.grid()

ax3.set_title("Cost")
ax3.plot(rendimiento["Proc."], rendimiento["Cost"], marker='o', mec = 'r',linestyle='-')
ax3.set_xlabel("Processors")
ax3.set_ylabel("Cost")
ax3.grid()


plt.show()
```

## Obtenci√≥n de valores relativos

```python
rendimiento_rt = pd.read_csv("datos.csv")

p_prima = 16

rendimiento_rt["Speedup"] = [p_prima / t for t in rendimiento["Time [s]"].values]
rendimiento_rt["Efficiency"] = [p_prima*s / p for p, s in zip(rendimiento["Proc."].values, rendimiento["Speedup"].values)]
rendimiento_rt["Max.acel.rel"] = [p/p_prima for p in rendimiento["Proc."].values]
```

```python
rendimiento_rt
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
fig.set_size_inches(19, 7)
                    
ax1.set_title("Reltative Speedup")
ax1.plot(rendimiento_rt["Proc."], rendimiento_rt["Speedup"], marker='*', mec = 'r',linestyle='-')
ax1.set_xlabel("Processors")
ax1.set_ylabel("Reltative Speedup")
ax1.grid()

ax2.set_title("Relative Efficiency")
ax2.plot(rendimiento_rt["Proc."], rendimiento_rt["Efficiency"], marker='*', mec = 'r',linestyle='-')
ax2.set_xlabel("Processors")
ax2.set_ylabel("Relative Efficiency")
ax2.grid()

ax3.set_title("Maxima Aceleraci√≥n Relativa")
ax3.plot(rendimiento_rt["Proc."], rendimiento_rt["Max.acel.rel"], marker='*', mec = 'r',linestyle='-')
ax3.set_xlabel("Processors")
ax3.set_ylabel("Maxima Aceleraci√≥n Relativa")
ax3.grid()


plt.show()
```
